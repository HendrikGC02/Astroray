```python
import sys
import matplotlib.pyplot as plt
sys.path.append(r'C:\Users\hcombrinck\OneDrive\Astroray\raytracer_addon\build\Release')

import raytracer_blender
```

```python
renderer = raytracer_blender.Renderer()
print("Module version:", raytracer_blender.__version__)
```

```python
# Create a simple scene
renderer = raytracer_blender.Renderer()

# Add a red sphere
renderer.add_sphere(
    center=[0.0, 0.0, 0.0],
    radius=1.0,
    material_type="metal",
    color=[0.8, 0.2, 0.2]
)

renderer.add_sphere(
    center=[-2.0, 2.0, -1.0],
    radius = 0.5,
    material_type="emmission",
    color=[50.0, 40.0, 40.0],
    param1=50.0  # intensity
)

# Add a ground plane (two triangles)
renderer.add_triangle(
    v0=[-5.0, -1.0, -5.0],
    v1=[5.0, -1.0, -5.0],
    v2=[5.0, -1.0, 5.0],
    material_type="diffuse",
    color=[0.5, 0.5, 0.5]
)

renderer.add_triangle(
    v0=[-5.0, -1.0, -5.0],
    v1=[5.0, -1.0, 5.0],
    v2=[-5.0, -1.0, 5.0],
    material_type="diffuse",
    color=[0.5, 0.5, 0.5]
)

# Setup camera
renderer.setup_camera(
    look_from=[0.0, 2.0, 5.0],
    look_at=[0.0, 0.0, 0.0],
    vup=[0.0, -1.0, 0.0],
    vfov=40.0,
    aspect_ratio=16/9,
    aperture=0.0,
    focus_dist=5.0,
    width=800,
    height=450
)

print("Rendering...")
pixels = renderer.render(samples_per_pixel=256, max_depth=4)

# Display
plt.figure(figsize=(12, 7))
plt.imshow(pixels)
plt.axis('off')
#plt.title('Raytraced Scene')
#plt.show()
```
